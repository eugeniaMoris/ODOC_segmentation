Dado que el recorte del fov no da una efectividad del 100% y se se pueda perder informacion de la imagen en el recorte. Se separaran los datos de entrenamiento y validacion de los de test.
Los datos de test ya no poseeran un recorte de fov y se almacenaran sin recortes.
Posteriormente para achicar las imagees de test se buscara un rescalado donde el radio resulltande de la imagen de test sea de un target radio (por el momento 255, es decir, 512/2)

**se debera** 

- Guardar las imagenes de test en una carpeta aparte 
- Al levantarlas (DataModule) aplicar un reescalado teniendo en cuenta el radio objetivo/target **(O ver de hacerlo antes)**

```python
def rescale_test(fundus_img,y,target_radio):
    lim_x_inf,lim_x_sup, lim_y_inf,lim_y_sup, radio = utils.crop_fov_limits(fundus_img)
    
    plt.imshow(fundus_img)
    plt.show()
    
    cut_img = fundus_img[lim_x_inf:lim_x_sup, lim_y_inf:lim_y_sup,:]
    plt.imshow(cut_img)
    plt.show()
    
    width, height,_ = fundus_img.shape    
    scale = target_radio / radio
    #scaledImg = F.resize(fundus_img, (scale*width,scale*height), interpolation=transforms.InterpolationMode.NEAREST)
    #fundus_img.resize((scale*width,scale*height))
    new_width = int(scale*width)
    new_height = int(scale*height)
    
    
    #if new_width % 2 != 0:
    #    new_width= new_width - 1
    #if new_height % 2 != 0:
    #    new_height= new_height - 1
        
    scaledImg = resize(fundus_img, (new_width,new_height))
    scaledCutImg = resize(cut_img, (512,512))
    
    
    
    size = max(new_width,new_height)
    squareImg = resize(scaledImg,(size,size))
    y_rescale = resize(y, (new_width,new_height))
    y_squared = resize(y_rescale,(size,size))
    
    
    print('Radio obtenido : ', radio, ' scala de la imagen: ', scale, '\n')
    print('tamao the la imagen cortada: ', scaledCutImg.shape)
    print('tamao de la imagen reescalada: ', scaledImg.shape)
    
    plt.imshow(scaledImg)
    plt.show()
    
    plt.imshow(squareImg)
    plt.show()
    
    print('y shape: ', y.shape)
    y =  y[lim_x_inf:lim_x_sup, lim_y_inf:lim_y_sup]
    print('y shape: ', y.shape)
    y = resize(y, (512,512))
    print('y shape: ', y.shape)
    
    return scaledImg,squareImg,scaledCutImg,y_rescale, y, y_squared
    
```

```python
scaled, squar,cut_img,y_r,y,y_q = rescale_test(test_img,test_mask,255)
```

```python
scaled = squar

loss = nn.CrossEntropyLoss()
config_path = '/mnt/Almacenamiento/ODOC_segmentation/codigo/config_files/DRISHTI_augm_f-75_p-25.ini'
config = ConfigParser()
config.read(config_path)

model = Unet(config, loss, model_name='prueba')

path_best = '/mnt/Almacenamiento/ODOC_segmentation/codigo/lightning_logs/drishti_model/matrix/version_19/checkpoints/DRISHTI_augm_f-75_p-25-epoch=109.ckpt'
print('Scaled shape: ', scaled.shape)

#CARGO MODELO
model = model.load_from_checkpoint(path_best, config= config, loss=loss, model_name='prueba')

#TRANSFORMO LA IMAGENE EN TENSOR
transform = transforms.ToTensor()
img = transform(scaled)
img= img.float()
shape = img.size()

#NORMALIZO
mean_r = torch.mean(img[0,:,:])
mean_g = torch.mean(img[1,:,:])
mean_b = torch.mean(img[2,:,:])

std_r = torch.std(img[0,:,:])+1e-6
std_g = torch.std(img[1,:,:])+1e-6
std_b = torch.std(img[2,:,:])+1e-6

img = F.normalize(img, [mean_r, mean_g, mean_b], [std_r,std_g, std_b]) #normalizo la imagen

#AGREGO LA DIMENSION BATCH SIZE
img= torch.unsqueeze(img, 0)

print('Scaled shape: ', img.shape, type(img))
#ARREGLAR LAS DIMENSIONES PARA QUE QUEDE EN 1, 3, W, H
y_hat = model.forward(img)
```

```python
def get_OCOD_crop(mask,delta):
    blobs_labels = measure.label(mask, background=0)
    region = measure.regionprops(blobs_labels)[0] 
    
    coordinates = region.bbox #(min_row, min_col, max_row, max_col)
    sub_h = coordinates[3]-coordinates[1]
    sub_w = coordinates[2]-coordinates[0]
    x,y = region.centroid
    if sub_h > sub_w: #REVISAR LOS LIMITES
        c1 = int(coordinates[1] - coordinates[1] * delta)
        c3 = int(coordinates[3] + coordinates[3] * delta)
        new_h = c3 - c1 
        if (new_h % 2) != 0:
            new_h = new_h + 1
            c3 = c3 + 1
        new_w = new_h
        half = new_w / 2
        c0 = y - half
        c2 = c2 + half
    else: #HACER PARA CUANDO ES MAS ANCHO QUE ALTO -- REVISAR LIMITES
        c1 = int(coordinates[1] - coordinates[1] * alpha)
        c3 = int(coordinates[3] + coordinates[3] * alpha)
        new_h = c3 - c1 
        if (new_h % 2) != 0:
            new_h = new_h + 1
            c3 = c3 + 1
        new_w = new_h
        half = new_w / 2
        c0 = y - half
        c2 = c2 + half
        
    
```
